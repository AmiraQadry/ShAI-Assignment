```python
import pandas as pd
import matplotlib.pyplot as plt
```

1. **Basic Data Exploration**: Identify the number of rows and columns in the dataset, determine the data types of each column, and check for missing values in each column.

```python
df = pd.read_csv('/content/Salaries.csv')

num_rows, num_columns = df.shape

data_types = df.dtypes

missing_values = df.isnull().sum()

print(f"Number of rows: {num_rows}, Number of columns: {num_columns}")
print("\nData types of each column:")
print(data_types)
print("\nMissing values in each column:")
print(missing_values)
```

2. **Descriptive Statistics**: Calculate basic statistics mean, median, mode, minimum, and maximum salary, determine the range of salaries, and find the standard deviation.

```python
salary_stats = df['TotalPay'].describe()

salary_mode = df['TotalPay'].mode().values[0]

salary_range = df['TotalPay'].max() - df['TotalPay'].min()

print("\nDescriptive Statistics for TotalPay:")
print(salary_stats)
print(f"\nMode Salary: {salary_mode}")
print(f"\nRange of salaries: {salary_range}")
```

3. **Data Cleaning**: Handle missing data by suitable method with explain why you use it.

```python
# Fill rows with missing values with mean
df.fillna(df.mean(), inplace=True)
```

4. **Basic Data Visualization**: Create histograms or bar charts to visualize the distribution of salaries, and use pie charts to represent the proportion of employees in different departments.

```python
# Histogram for distribution of salaries
plt.hist(df['TotalPay'], bins=20, edgecolor='black')
plt.title('Salary Distribution')
plt.xlabel('TotalPay')
plt.ylabel('Frequency')
plt.show()

# Pie chart represents the proportion of employees in different departments
department_counts = df['JobTitle'].value_counts().head(10)
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 10 Job Titles Distribution')
plt.show()
```

5. **Grouped Analysis**: Group the data by one or more columns and calculate summary statistics for each group, and compare the average salaries across different groups.


```python
# Group data by 'Year' and calculate average salaries for each year
average_salary_by_year = df.groupby('Year')['TotalPay'].mean()

print("\nAverage Salary by Year:")
print(average_salary_by_year)
```

6. **Simple Correlation Analysis**: Identify any correlation between salary and another numerical column, and plot a scatter plot to visualize the relationship.

```python
correlation = df['TotalPay'].corr(df['OtherPay'])

plt.scatter(df['TotalPay'], df['OtherPay'])
plt.title('Scatter Plot: TotalPay vs OtherPay')
plt.xlabel('TotalPay')
plt.ylabel('OtherPay')
plt.show()

print(f"\nCorrelation between TotalPay and OtherPay: {correlation}")
```
